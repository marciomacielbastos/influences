```python
import numpy as np
from collections import Counter
from scipy.cluster.vq import kmeans,kmeans2
import pylab as pl
import IPython
import datetime as dt
import codecs
import pandas as pd
import networkx as nx
import datetime as dt
import matplotlib
import matplotlib.pyplot as plt
import operator
%pylab inline 
```

```python
datapath='../MMD/datasets/influences.csv'
s = codecs.open(datapath, encoding="utf-8").readlines()
lines = [line.strip() for line in s]
lines2 = []
for line in lines:
    line = line.replace('",','";')
    line = line.replace('"','').strip()
    lines2.append(line)
dataset = np.array([(line.split(';')[0],line.split(';')[1]) for line in lines2])[1:]
influenciadores = np.array(dataset[:,0])
influenciados = np.array(dataset[:,1])
pessoas = np.union1d(influenciadores,influenciados)
```

```python
total=pd.DataFrame(dataset)
```

```python
total.info()
```

```python
total = total.rename_axis({0:'Influenciador', 1:'Influenciado'}, axis=1)
total
```

```python
top_10_influenciadores = total.groupby('Influenciador').Influenciado.count().order(ascending = False)[:10]
top_10_influenciadores
```

```python
top_10_influenciados = total.groupby('Influenciado').Influenciador.count().order(ascending = False)[:10]
top_10_influenciados
```

```python
todos_influenciadores = total.groupby('Influenciador').Influenciado.count().index
todos_influenciadores
```

```python
todos_influenciados = total.groupby('Influenciado').Influenciador.count().index
todos_influenciados
```

```python
todos_os_filosofos = todos_influenciadores.union(todos_influenciados)
todos_os_filosofos
```

```python
G = nx.DiGraph()
G.add_nodes_from(todos_os_filosofos)
G.add_edges_from(np.array(total))
plt.figure(figsize=(16,16))
nx.draw(G, node_color='blue', node_size=20, with_labels=False)
```

```python
prank = nx.pagerank(G)
prank
```
