#### Questão 3: Quais são os 5 maiores "clusters" de personalidades que foram influenciados por um grupo semelhante de influentes? Considere um cluster como um número X de personalidades que foram influenciados pelos mesmos Y influenciadores, sendo Y um número arbitrário. Comecem com Y=2 e cresçam até que não haja nenhum grupo de X influenciados sendo influenciados por Y influentes. 

```python
influenciadores_ordenados = total.groupby('Influenciador').Influenciado.count().order(ascending = False)[:].index
influenciadores_ordenados
```

```python
clusters = [[],[],[],[],[]]
tamanhos = [0,0,0,0,0]
clusterizadores = [[],[],[],[],[]]
def acha_5_maiores_clusters(Y,conj=[], i=0):
    if Y>1:
        for j in range(i,len(influenciadores_ordenados)):
            inf = influenciadores_ordenados[j]
            if len(G.successors(inf))<min(tamanhos):
                break
            if len(list(set.intersection(*[set(G.successors(p)) for p in conj+[inf]])))<=min(tamanhos):
                continue
            acha_5_maiores_clusters(Y-1,conj+[inf], j+1)
    else:    
        for j in range(i,len(influenciadores_ordenados)):
            inf = influenciadores_ordenados[j]
            if len(G.successors(inf))<=min(tamanhos):
                break
            temp_cluster = list(set.intersection(*[set(G.successors(p)) for p in conj+[inf]]))
            if len(temp_cluster)>min(tamanhos):
                index = tamanhos.index(min(tamanhos))
                tamanhos[index]=len(temp_cluster)
                clusters[index]=temp_cluster
                clusterizadores[index]=conj+[inf]      
```
